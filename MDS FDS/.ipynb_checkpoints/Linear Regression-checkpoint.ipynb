#### Prepare Dataset

```python
np.random.seed(11)
x = np.random.randn(100)
```

```python
y = 2 * x + 5 + np.random.randn(100) * 0.9
```

#### Plot Dataset

```python
plt.figure(figsize=(20,8))
plt.scatter(x, y)
```

```python
x1 = (x-np.mean(x)/np.std(x))
```

```python
plt.figure(figsize=(20,8))
plt.scatter(x1,y)
```

#### Intialize Weight and Bias

```python
w = 1
b = 0

epocs = 100
alpha = 0.1
```

```python
def linear_regression(w,b):
    y_hat = w * x1 + b
    return y_hat
```

```python
def RMSE(y_hat, y):
    return np.sqrt(np.sum(y-y_hat))
```

#### Update Weight and Bais

```python
y_hat = linear_regression(w, b)
for epoc in range(epocs):
    loss = RMSE(y_hat, y)
    dw = np.sum(np.dot((y_hat-y),x1))/2
    db = np.sum(y_hat - y) / 2

    # update w and b
    w -= alpha * dw/(len(x1))
    b -=  alpha * db/(len(x1))

    y_hat = linear_regression(w, b)
    print(f"Loss in epoch {epoc+1}, {RMSE(y_hat,y)}")
```

#### Plot the Loss

```python
y_pred = linear_regression(w,b)
plt.scatter(x1,y)
plt.plot(x1,y_pred,color='red')
```

#### Z Scaling

```python
def scale(d):
    mean = d.mean()
    sd = d.std()

    return (d-mean)/sd
```

#### Min Batch

```python
def data_iter(batch_size,features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))

    np.random.shuffle(indices)
    # the example are read at random in no particular order
    for i in range(0, num_examples, batch_size):
        batch_indices = np.array(indices[i:min(i+batch_size,num_examples)])

        yield features[batch_indices], labels[batch_indices]
    
```

#### Simple Linear Regression Class From Scratch

```python
class LinearRegression():
    def __init__(self, X, y, W, b):
        self.X = X
        self.y = y
        self.W = W
        self.b = b
        self.y_pred = self.LinearModel()

    def LinearModel(self):
        return np.dot(self.X, self.W) + self.b

    def mse_loss(self, y, y_pred):
        return np.mean(np.sum(np.square(y-y_pred))) /2

    def train(self, num_epochs = 10, batch_size = 10):
        # training

        for epoc in range(num_epochs):
            for X, y in data_iter(batch_size, self.X, self.y):

                y_pred = self.predict(X)

                error = (y - y_pred)

                # update parameters

                grad_W = np.sum(error.T.dot(X))
                grad_b = np.sum(y - y_pred)

                # update weight and bais term using gradient
                self.W = self.W -  (1/batch_size * lr * grad_W)
                self.b = self.b - (1/batch_size * lr * grad_b)
            y_pred = self.predict(X)
            final_loss = self.mse_loss(y, y_pred) / batch_size
            print(f'epoch {epoc+1}, loss {float(final_loss):f}')

    def predict(self, X):
        return np.dot(X, self.W) + self.b
                
```

```python
def scale(d):
    mean = d.mean()
    sd = d.std()

    return (d - mean) / sd
```

```python
def data_iter(batch_size,features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))

    np.random.shuffle(indices)
    # the example are read at random in no particular order
    for i in range(0, num_examples, batch_size):
        batch_indices = np.array(indices[i:min(i+batch_size,num_examples)])

        yield features[batch_indices], labels[batch_indices]
    
```

```python
def load_dataset():
    data = load_diabetes()
    X , y = data.data, data.target
    return X, y
```

```python
class LogisticRegression():
    """ This the scratch implementation of Logistic Regression"""

    def __init__(self, X, y):
        self.params = {}
        self.m, self.n = X.shape
        self.params['W'] = np.random.randn(self.n,1) * 0.001
        self.params['b'] = np.zeros(1)

        self.X = X
        self.y = y
        self.result = pd.DataFrame()

    def train(self, alpha = 0.001, epochs = 10):
        for epoch in range(epochs):
            print("Epoch:", epoc, end="")
            z = np.dot(self.X,self.params['W']+self.params['b'])

            self.y_pred = self.sigmoid(z)
            self.result[0] = self.y

            # update the parameters
            self.params['W'] = self.params['W'] - alpha * 1/self.m * np.dot(self.X.T,(self.y_pred-np.reshape(self.y,(self.m,1))))
            self.params['b'] = self.params['b'] - alpha * 1/self.m * np.sum(self.y_pred-np.reshape(self.y,(self.m,1)))

            self.y_pred = self.sigmoid(np.dot(self.X,self.params['W'])+self.params['b'])
            loss = self.loss(self.y, self.y_pred)

            self.result[1] = self.y_pred
            print(",loss=",loss)

        print("Final loss = ", loss)
        print(f"W: {self.params['W']}, b: {self.params['b']}")

    def loss(self, y, y_pred):
        y_zero_loss = y.T.dot(np.log(y_pred))
        y_one_loss = (1-y).T.dot(np.log(1-y_pred))

        return -np.sum(y_zero_loss+y_one_loss)/len(y)

    def sigmoid(self, z):
        return 1.0 / (1 + np.exp(-z))

    def predict(self, X):
        return self.sigmoid(np.dot(X, self.params['W'])+self.params['b'])
```

```python
X , y = load_dataset()
```

```python
alpha = 0.0001
epochs = 300
log_model = LogisticRegression(X,y)
log_model.train(alpha, epochs)
```
